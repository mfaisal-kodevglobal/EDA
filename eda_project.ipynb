```python
print('Missing Values')

## Here we will check the percentage of nan values present in each feature
## 1 -step make the list of features which has missing values
features_with_na=[features for features in df1.columns if df1[features].isnull().sum()>1]
## 2- step print the feature name and the percentage of missing values

for feature in features_with_na:
    print(feature, np.round(df1[feature].isnull().mean(), 4),  ' % missing values')
```

```python
print('Numerical Variables')
#list of numerical variables

numerical_features = [feature for feature in df1.columns if df1[feature].dtypes != 'O']

print('Number of numerical variables: ', len(numerical_features))

#visualising numerical variables
df1[numerical_features].head()
```

```python
print('Temporal Variables(i.e. Datetime variables)')
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]

year_feature
```

```python
print('Categorical variables')
categorical_features=[feature for feature in df1.columns if df1[feature].dtypes== 'O']

categorical_features
```

```python
df1[categorical_features].head()
```

```python

for feature in categorical_features:
    print("The feature is {} and number of categories are {}".format(feature,len(df1[feature].unique())))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
# sns.barplot(df1['Gender'])
```

```python
print('Now we will try to find out the relationship between categorical variables and dependent features')
for feature in categorical_features:
    data=df1.copy()
    data.groupby(feature)['BMI'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('BMI')
    plt.title(feature)
    plt.show()
```

```python
print('Now we will try to find out the relationship between categorical variables and dependent features')
for feature in categorical_features:
    data=df1.copy()
    data.groupby(feature)['Systolic BP'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('Systolic BP')
    plt.title(feature)
    plt.show()
```

```python
# Diastolic BP
print('Now we will try to find out the relationship between categorical variables and dependent features')
for feature in categorical_features:
    data=df1.copy()
    data.groupby(feature)['Diastolic BP'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('Diastolic BP')
    plt.title(feature)
    plt.show()
```

```python
# Data show the Glucose Level

# Create a bar plot
plt.bar(df1['Age'], df1['Glucose'])

# Add labels and title
plt.xlabel('Age[Both Male and Female]')
plt.ylabel('Glucose')
plt.title('Number of Patients with Diabetes by Type')

# Show the plot
plt.show()
```
